BIVARIATE AND MULTIVARIATE ANALYSIS

1. Correlation Heatmap

Purpose: To visualize the correlation between numerical variables.

Plot Type: Heatmap

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Now create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```

2. Pairplot

Purpose: To observe pairwise relationships and distributions for all numerical features.

Plot Type: Pairplot

```python
sns.pairplot(df, hue='booking_complete')  # Use 'hue' for categorical separation if relevant
plt.title('Pairplot of Numerical Features')
plt.show()
```

3. Scatter Plot

Purpose: To analyze the relationship between two continuous variables.

Plot Type: Scatter plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='purchase_lead', y='flight_duration', hue='booking_complete', data=df, palette='coolwarm')
plt.title('Purchase Lead vs Flight Duration')
plt.show()
```

4. Box Plot

Purpose: To compare distributions of a continuous variable across categories.

Plot Type: Box plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='trip_type', y='purchase_lead', data=df)
plt.title('Purchase Lead Time by Trip Type')
plt.show()
```

5. Violin Plot

Purpose: To compare the distribution and probability density of a continuous variable across categories.

Plot Type: Violin plot

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='flight_day', y='flight_duration', hue='booking_complete', data=df, split=True, palette='muted')
plt.title('Flight Duration by Day of Week')
plt.show()
```

6. Bar Plot (for Aggregates)

Purpose: To visualize the mean or sum of a numerical variable grouped by categories.

Plot Type: Bar plot

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='flight_hour', y='num_passengers', hue='booking_complete', data=df, ci=None, palette='viridis')
plt.title('Average Number of Passengers by Flight Hour')
plt.show()
```

7. FacetGrid

Purpose: To create multiple plots to visualize how data distributions vary across subsets.

Plot Type: FacetGrid

```python
g = sns.FacetGrid(df, col='trip_type', hue='booking_complete', height=4, aspect=1.2)
g.map(sns.histplot, 'purchase_lead', kde=True).add_legend()
g.fig.suptitle('Purchase Lead Distribution by Trip Type and Booking Completion', y=1.05)
plt.show()
```

8. Count Plot (for categorical comparisons)

Purpose: To compare counts of observations across different categories.

Plot Type: Count plot

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='flight_day', hue='booking_complete', data=df, palette='cubehelix')
plt.title('Booking Completion by Flight Day')
plt.show()
```

9. Multivariate Plotting using Seaborn's catplot

Purpose: To analyze relationships between multiple variables across different categories.

Plot Type: Catplot

```python
sns.catplot(x='flight_day', y='flight_duration', hue='trip_type', col='booking_complete', kind='box', data=df, height=5, aspect=1.2)
plt.suptitle('Flight Duration by Day, Trip Type, and Booking Completion', y=1.05)
plt.show()
```

10. Swarm Plot

Purpose: To visualize the distribution of a categorical variable along with the distribution of a continuous variable.

Plot Type: Swarm plot

```python
plt.figure(figsize=(12, 6))
sns.swarmplot(x='trip_type', y='length_of_stay', hue='booking_complete', data=df, palette='deep')
plt.title('Length of Stay by Trip Type and Booking Completion')
plt.show()
```
